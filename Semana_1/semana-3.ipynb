En Python sabemos que una función puede llamar a otra función, esto incluso es posible para las funciones que se llaman a si mismas.

````python
def suma(a,b):
    print(a+b)
    return a + b
    

def capture_numeros()->None:
    a = int(input("numero a: "))
    b = int(input("numero b: "))
    return suma(a,b) + 100

def menu() -> None:
    print("1. suma \n 2. resta \n 3. division")

# reconoce los metodos
## llame a las funciones
menu()
print(f"el resultado de la operacion es: {capture_numeros()}")
````

````python
def factorial(x):
    """This is a recursive function
    to find the factorial of an integer"""

    if x == 1:
        return 1
    else:
        return (x * factorial(x-1)) # 3 * 2 * 1


num = 3
print("The factorial of", num, "is", factorial(num))
````

#### Ejemplo clase persona

````python
class persona:
    # constructor
    def __init__(self,nombre: str, edad: int, altura: float, lista_hobbies: list) -> None:
        '''
        Constructor que inicializa la clase
        '''        
        self.nombre  = nombre
        self.edad    = edad
        self.altura  = altura
        self.hobbie  = lista_hobbies
    
    # obtener datos de persona
    def get_altura(self) -> float:
        return self.altura
    
    def get_name(self):
        return self.nombre
    
    def get_edad(self):
        return self.edad
        
    def set_edad(self,nueva_edad: int)->None:
        self.edad = nueva_edad
````

````python
### Conocer como llamamos un clase ?

tatiana = persona("tatiana",28,1.70,["natacion","futbol","rugby subacutico","water polo"])
ander = persona("Ander",23,1.60,["natacion","futbol"])
````

````python
print(f"la persona se llama {tatiana.nombre} su edad es {tatiana.edad}")
print(f"la persona se llama {ander.nombre} su edad es {ander.edad}")
````

````python
print(f"edad de ander {ander.get_edad()}")
ander.set_edad(24)
print(f"nueva edad de ander {ander.get_edad()}")
````
