#### Python recursive function

En Python sabemos que una función puede llamar a otra función, esto incluso es posible para las funciones que se llaman a si mismas.

````python
def factorial(x):
    """This is a recursive function
    to find the factorial of an integer"""

    if x == 1:
        return 1
    else:
        return (x * factorial(x-1))


num = 3
print("The factorial of", num, "is", factorial(num))
````

Una recursión termina cuando el número base se reduce a 1, incluso la función debe tener una condición base para detenerse o sino la función cae en un bucle infinito.
