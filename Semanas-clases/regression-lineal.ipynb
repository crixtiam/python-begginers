# Clase: Introducción a la Regresión

## 1. ¿Qué es la Regresión?
La regresión es un método estadístico utilizado para modelar la relación entre una variable dependiente (**Y**) y una o más variables independientes (**X**).

### Tipos de Regresión
- **Regresión Lineal:** Modelo que asume una relación lineal entre las variables.
- **Regresión Polinómica:** Extensión de la regresión lineal que utiliza términos polinómicos.
- **Regresión Logística:** Utilizada para variables dependientes categóricas.

---

## 2. Modelo de Regresión Lineal
El modelo básico es:
\[
y = ax + b + e
\]
Donde:
- \( y \): Variable dependiente (respuesta).
- \( x \): Variable independiente (predictora).
- \( a \): Pendiente de la línea de regresión.
- \( b \): Intercepto o constante.
- \( e \): Error o término residual.

### Suposiciones del Modelo
1. Linealidad: La relación entre \( X \) e \( Y \) es lineal.
2. Independencia: Los errores son independientes entre sí.
3. Homocedasticidad: La varianza de los errores es constante.
4. Normalidad: Los errores siguen una distribución normal.

---

## 3. Ejemplo de Interpretación
Dado el modelo:
\[
y = 2x + 3
\]
- Por cada unidad que aumenta \( x \), \( y \) aumenta en 2 unidades.
- Cuando \( x = 0 \), \( y = 3 \) (el intercepto).

---

## 4. Métricas para Evaluar el Modelo
- **Coeficiente de Determinación (\( R^2 \)):** Indica qué porcentaje de la variabilidad en \( Y \) es explicado por el modelo.
- **Error Cuadrático Medio (MSE):** Mide la diferencia promedio al cuadrado entre los valores observados y los predichos.

---

## 5. Aplicaciones
- **Economía:** Predecir ventas en función de precios.
- **Medicina:** Relación entre dosis de medicamentos y efectos.
- **Ingeniería:** Modelado de fallos en sistemas mecánicos.

---

## 6. Conclusión
La regresión es una herramienta poderosa para identificar relaciones y hacer predicciones. Es fundamental entender sus supuestos y limitaciones para aplicar los modelos correctamente.

# Ejemplo: Ajuste de una Regresión Lineal

## 1. Conjunto de Datos
Dado el siguiente conjunto de observaciones:
$$
\{(X_1, Y_1), (X_2, Y_2), (X_3, Y_3), (X_4, Y_4), (X_5, Y_5)\} = \{(1, 2), (2, 4), (3, 5), (4, 4), (5, 5)\}
$$

## 2. Cálculo de Promedios
Primero, calculamos los promedios de \( X \) y \( Y \):
$$
\bar{X} = \frac{1 + 2 + 3 + 4 + 5}{5} = 3, \quad \bar{Y} = \frac{2 + 4 + 5 + 4 + 5}{5} = 4
$$

## 3. Fórmulas de los Coeficientes
### a. Pendiente (\( a \)):
$$
a = \frac{\sum_{i=1}^n (X_i - \bar{X})(Y_i - \bar{Y})}{\sum_{i=1}^n (X_i - \bar{X})^2}
$$
Calculamos cada término:
- $X_i - \bar{X} = \{-2, -1, 0, 1, 2\} $
- $Y_i - \bar{Y} = \{-2, 0, 1, 0, 1\}$
- $Producto: (X_i - \bar{X})(Y_i - \bar{Y}) = \{4, 0, 0, 0, 2\}$
- $Suma de productos: \sum (X_i - \bar{X})(Y_i - \bar{Y}) = 6$
- $(X_i - \bar{X})^2 = \{4, 1, 0, 1, 4\} $
- $Suma de cuadrados:  \sum (X_i - \bar{X})^2 = 10 $

Sustituimos:
$$
a = \frac{6}{10} = 0.6
$$

### b. Intercepto (\( b \)):
$$
b = \bar{Y} - a\bar{X}
$$
$$
b = 4 - (0.6)(3) = 4 - 1.8 = 2.2
$$

### Modelo Ajustado:
$$
Y = 0.6X + 2.2
$$

## 4. Graficación
Podemos graficar los puntos y la línea ajustada para visualizar el modelo.

### Código Python (Opcional para Graficar):

# Evaluación del Modelo de Regresión

## 1. Coeficiente de Determinación (\( R^2 \))
El coeficiente de determinación $R^2$ mide qué tan bien el modelo ajustado explica la variabilidad de los datos. Se calcula como:

$$
R^2 = 1 - \frac{\text{Suma de los errores al cuadrado (SSE)}}{\text{Suma total de cuadrados (SST)}}
$$

Donde:
- **SSE**: Suma de los errores al cuadrado.
$$
SSE = \sum_{i=1}^n (Y_i - Y_{\text{predicho}})^2
$$
- **SST**: Suma total de cuadrados.
$$
SST = \sum_{i=1}^n (Y_i - \bar{Y})^2
$$

---

## 2. Cálculo de $R^2$ para el Ejemplo
### Valores Observados y Predichos
Modelo ajustado: $Y = 0.6X + 2.2$

| $X$ | $Y_{\text{observado}}$ | $Y_{\text{predicho}}$ | $(Y_{\text{observado}} - Y_{\text{predicho}})^2$ |
|-----|------------------------|-----------------------|--------------------------------------|
| 1   | 2                      | $0.6(1) + 2.2 = 2.8$ | $(2 - 2.8)^2 = 0.64$                 |
| 2   | 4                      | $0.6(2) + 2.2 = 3.4$ | $(4 - 3.4)^2 = 0.36$                 |
| 3   | 5                      | $0.6(3) + 2.2 = 4.0$ | $(5 - 4.0)^2 = 1.0$                  |
| 4   | 4                      | $0.6(4) + 2.2 = 4.6$ | $(4 - 4.6)^2 = 0.36$                 |
| 5   | 5                      | $0.6(5) + 2.2 = 5.2$ | $(5 - 5.2)^2 = 0.04$                 |

- $SSE = 0.64 + 0.36 + 1.0 + 0.36 + 0.04 = 2.4$

### Suma Total de Cuadrados ($SST$):
$$
SST = \sum_{i=1}^n (Y_i - \bar{Y})^2
$$
- $\bar{Y} = 4$
- $(Y_1 - \bar{Y})^2 = (2 - 4)^2 = 4$
- $(Y_2 - \bar{Y})^2 = (4 - 4)^2 = 0$
- $(Y_3 - \bar{Y})^2 = (5 - 4)^2 = 1$
- $(Y_4 - \bar{Y})^2 = (4 - 4)^2 = 0$
- $(Y_5 - \bar{Y})^2 = (5 - 4)^2 = 1$

$$
SST = 4 + 0 + 1 + 0 + 1 = 6
$$

### Cálculo de $R^2$:
$$
R^2 = 1 - \frac{SSE}{SST}
$$

$$
R^2 = 1 - \frac{2.4}{6} = 1 - 0.4 = 0.6
$$

---

## 3. Interpretación de $R^2$
El valor de $R^2 = 0.6$ indica que el modelo explica el 60% de la variabilidad de los datos observados. Aunque el modelo tiene un ajuste razonable, aún existe un 40% de la variabilidad que no está explicada, lo que podría atribuirse a factores no incluidos en el modelo o errores aleatorios.

---

## 4. Conclusión
El modelo ajustado:
$$
Y = 0.6X + 2.2
$$
es útil para describir la relación entre $X$ e $Y$, pero podría mejorarse si se consideran variables adicionales o modelos más complejos.


````python
import matplotlib.pyplot as plt
import numpy as np

X = [1, 2, 3, 4, 5]
Y = [2, 4, 5, 4, 5]

a = 0.6
b = 2.2
X_line = np.linspace(1, 5, 100)
Y_line = a * X_line + b

plt.scatter(X, Y, color='blue', label='Datos observados')
plt.plot(X_line, Y_line, color='red', label='Línea ajustada: Y = 0.6X + 2.2')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Regresión Lineal')
plt.legend()
plt.grid()
plt.show()
````

````python
import numpy as np

X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])

X_mean = np.mean(X)
Y_mean = np.mean(Y)

numerator = np.sum((X - X_mean) * (Y - Y_mean))
denominator = np.sum((X - X_mean) ** 2)
a = numerator / denominator

b = Y_mean - a * X_mean

print(f"Modelo ajustado: Y = {a:.2f}X + {b:.2f}")
````

# Fórmulas de Covarianza y Varianza y su Aplicación en la Regresión Lineal

La **covarianza** y la **varianza** son dos conceptos fundamentales en estadística que se utilizan para entender la relación entre variables y la dispersión de los datos, respectivamente. Estos conceptos son muy útiles en el contexto de la **regresión lineal**, ya que ayudan a calcular los coeficientes del modelo, como la pendiente (\( a \)) y el intercepto (\( b \)).

## 1. Varianza

La **varianza** mide la dispersión o la cantidad de variabilidad en una variable. Indica cuánto varían los valores de una variable respecto a su media.

La **varianza** de una variable \( X \) se calcula con la siguiente fórmula:

$$
\text{Var}(X) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})^2
$$

Donde:
- $ X_i $ son los valores de la variable  X ,
- $ \bar{X} $ es el valor promedio de la variable X ,
- $ n $ es el número de observaciones.

La **varianza** mide cuán dispersos están los datos de \( X \) alrededor de su media \( \bar{X} \). Cuanto mayor sea la varianza, más dispersos están los datos.

## 2. Covarianza

La **covarianza** mide la relación lineal entre dos variables \( X \) e \( Y \). Indica cómo varían juntas las dos variables. Si las dos variables tienden a aumentar o disminuir juntas, la covarianza será positiva. Si una variable aumenta mientras la otra disminuye, la covarianza será negativa. Si no hay una relación lineal clara, la covarianza será cercana a cero.

La **covarianza** entre dos variables \( X \) e \( Y \) se calcula con la siguiente fórmula:

$$
\text{Cov}(X, Y) = \frac{1}{n} \sum_{i=1}^{n} (X_i - \bar{X})(Y_i - \bar{Y})
$$

Donde:
- $X_i$ y  $Y_i$ son los valores de las variables $ X $ e $ Y $,
- $ \bar{X}$  y  $\bar{Y}$  son los valores promedio de las variables $ X $ e $ Y $,
- \( n \) es el número de observaciones.

La **covarianza** nos dice si las dos variables tienen una relación lineal positiva o negativa, pero no nos da una medida de la fuerza de esta relación. Para esto, usamos el **coeficiente de correlación**.

## 3. Uso en Regresión Lineal

En el contexto de la regresión lineal, tanto la **covarianza** como la **varianza** son esenciales para calcular la pendiente \( a \) de la recta de regresión.

### Cálculo de la pendiente (\( a \)) de la recta de regresión

La pendiente \( a \) de la línea de regresión se puede expresar en términos de la **covarianza** y la **varianza** de las variables \( X \) y \( Y \) de la siguiente manera:

$
a = \frac{\text{Cov}(X, Y)}{\text{Var}(X)}
$

Donde:
- $ \text{Cov}(X, Y) $ es la covarianza entre las variables X  e  Y ,
- $ \text{Var}(X) $ es la varianza de la variable  X .

### Cálculo del intercepto (\( b \)) de la recta de regresión

Una vez que tenemos la pendiente \( a \), el **intercepto** \( b \) de la recta de regresión se puede calcular usando la fórmula:

$$
b = \bar{Y} - a\bar{X}
$$

## 4. Ejemplo en el Contexto de la Regresión Lineal

Supón que tenemos los siguientes puntos de datos:

$$
\{(1, 2), (2, 4), (3, 5), (4, 4), (5, 5)\}
$$

1. **Cálculo de la varianza de \( X \)**:
   - Calculamos el promedio $ \bar{X}$ de  X ,
   - Luego, usamos la fórmula de varianza para calcular $ \text{Var}(X) $.

2. **Cálculo de la covarianza entre \( X \) y \( Y \)**:
   - Calculamos el promedio $\bar{Y}$ de \( Y \),
   - Usamos la fórmula de covarianza para calcular $ \text{Cov}(X, Y) $.

3. **Cálculo de la pendiente \( a \)**:
   Usamos la fórmula de la pendiente:

   $
   a = \frac{\text{Cov}(X, Y)}{\text{Var}(X)}
   $

4. **Cálculo del intercepto \( b \)**:
   Usamos la fórmula del intercepto:

   $
   b = \bar{Y} - a\bar{X}
   $

## 5. Resumen de la Relación con la Regresión Lineal

- **Varianza**: Mide la dispersión de los datos de la variable independiente \( X \).
- **Covarianza**: Mide cómo varían juntas las variables \( X \) e \( Y \).
- En la regresión lineal, la **pendiente** de la recta (\( a \)) se calcula usando la **covarianza** y la **varianza** de \( X \) y \( Y \).
